```python
import json
def strip_stream(filename='APEX-CMZ_ReductionTests.ipynb'):
    jay = json.load(open(filename))
    for cell in jay['worksheets'][0]['cells']:
        if 'outputs' in cell:
            cell['outputs'] = [x for x in cell['outputs'] if not (x['output_type'] == 'stream'
                                                                  and '\r' in x['text'])]
    with open(filename,'w') as outf:
        json.dump(jay, outf)
    return jay
```
