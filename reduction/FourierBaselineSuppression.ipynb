Fourier Baseline Suppression
============================

The large-scale features that are obvious to the eye are large amplitude and spatially variable.  These cannot easily be tackled with fourier-based suppression (I tried a lot of different approaches) but they are smooth enough that local polynomials handle them well.

The local, low-amplitude, high-frequency modes are much more problematic.  I discovered upon deeper inspection that the 2013 data is *much* more severely affected than the 2014 data.  Therefore, it may be possible to supress the bad modes in these data without dramatic negative effects.

It helps that the 2013 data did not include 13CO, so we don't need to worry about losing bright lines.

```python
from pyspeckit.spectrum.readers import read_class
cl = read_class.ClassObject('/Volumes/passport/apex/raw/dotapex/M-091.F-0019-2013-2013-06-08.apex')
```

```python
spectra = cl.get_spectra(sourcere='^SGRA', telescope=list(cl.tels)[0])
data = np.array([s for s,h in spectra])
```

```python
# this takes about 90s
allsuppressed = np.array([suppress_frange(data[ii,:], frange=[1.6e-3, 1.9e-3], width=2) for ii in range(data.shape[0])])
```

```python
import pylab as pl
import matplotlib
import paths
matplotlib.rc_file(paths.pcpath('pubfiguresrc'))
```

```python
# before: show the mean *absolute value* spectrum in order to avoid
absmeanspec = abs(data).mean(axis=0)
pl.plot(absmeanspec)
pl.ylim(2.35, 2.66)
```

```python
# after
absmeanspec_suppr = np.abs(allsuppressed).mean(axis=0)
pl.plot(absmeanspec_suppr)
pl.ylim(2.35, 2.66)
```

```python
# diff
pl.plot(absmeanspec-absmeanspec_suppr)
```

This looks like a dramatic success to me.  There's no sign of localized structure removal that would indicate a real line being suppressed, and there's a clear and huge improvement in the high-frequency baseline components throughout.

The range suppressed is 27.5-32.5 km/s.

```python
# Demonstrate the difference in the fourier spectra
ft = np.fft.fft(absmeanspec)
fts = np.fft.fft(absmeanspec_suppr)
freq = np.fft.fftfreq(ft.size)
pl.loglog(freq, abs(ft), color='k', linewidth=2)
pl.loglog(freq, abs(fts), color='r', linewidth=1)
pl.xlim(freq.min(), freq.max())
pl.ylim(1e1,1e3)
```

This is actually a bit surprising - apparently there is a significantly under-represented mode that we have scaled up?  In any case, the practical effects are solid.

We'll zoom in for a closer look, though...

```python
pl.semilogy(freq, abs(ft), color='k', linewidth=2)
pl.semilogy(freq, abs(fts), color='r', linewidth=1)
pl.xlim(freq.min(), freq.max())
pl.ylim(1e1,2e2)
pl.xlim(1e-3,2.5e-3)
```

```python
def do_fourier_plot(absmeanspec, absmeanspec_suppr, zoom=False):
    ft = np.fft.fft(absmeanspec)
    fts = np.fft.fft(absmeanspec_suppr)
    freq = np.fft.fftfreq(ft.size)
    pl.loglog(freq, abs(ft), color='k', linewidth=2)
    pl.loglog(freq, abs(fts), color='r', linewidth=1)
    pl.xlim(freq.min(), freq.max())
    pl.ylim(1e1,1e3)
    if zoom:
        pl.xlim(1e-3,2.5e-3)
        pl.gca().set_xscale('linear')
```

```python
def specplots(data, allsuppressed):
    pl.subplot(3,1,1)
    pl.plot(absmeanspec)
    pl.subplot(3,1,2)
    pl.plot(absmeanspec_suppr)
    pl.subplot(3,1,3)
    pl.plot(absmeanspec-absmeanspec_suppr)
```
